## Exploración inicial
Describe las columnas y analiza los tipos de datos.

```python
# Escribe tu código aquí para explorar el DataFrame
import pandas as pd

# Cargar datos
df_biomasa = pd.read_csv("resultado_reto3_Karen.csv")
df_clima = pd.read_csv("resultados_con_clima.csv")

# Combinar datos de biomasa con información de clima
datos_combinados = df_biomasa.merge(df_clima[['sitio', 'clima']], on='sitio', how='left')

# Análisis de biomasa por tipo de clima
print("ANÁLISIS DE BIOMASA POR TIPO DE CLIMA")

# Calcular promedio de biomasa por clima
biomasa_por_clima = datos_combinados.groupby('clima')['biomasa'].mean()
print("\nPromedio de biomasa por tipo de clima:")
for clima, promedio in biomasa_por_clima.items():
    print(f"{clima}: {promedio:.1f}")

# Contar sitios por categoría de biomasa y clima
print("\nDistribución de clasificación de biomasa por clima:")
distribucion = datos_combinados.groupby(['clima', 'Clasificación_biomasa']).size()
print(distribucion)

# Función para clasificar temperatura
def clasificar_temperatura(temp):
    if temp < 18:
        return 'Fría'
    elif temp <= 24:
        return 'Templada'
    else:
        return 'Cálida'

# Aplicar clasificación
datos_combinados['categoria_temp'] = datos_combinados['temperatura'].apply(clasificar_temperatura)
print("\nDistribución por categoría de temperatura:")
print(datos_combinados['categoria_temp'].value_counts())
```

## Desarrollo del reto
Aplica funciones y análisis según el objetivo. Usa `def` para definir tus funciones.

```python
# Crea aquí tus funciones y análisis
import pandas as pd

# Cargar datos de Santiago
df_luz = pd.read_csv("resultado_reto2_santiago_final.csv")

# Análisis de relación entre luz y biomasa
print("ANÁLISIS RELACIÓN LUZ-BIOMASA")

# Estadísticas por categoría de luz
print("\nEstadísticas de biomasa por categoría de luz:")
stats_luz = df_luz.groupby('categoria_luz')['Biomasa'].agg(['mean', 'min', 'max'])
print(stats_luz)

# Función para evaluar productividad
def evaluar_productividad(fila):
    relacion = fila['Biomasa'] / fila['Luz_Cantidad'] if fila['Luz_Cantidad'] > 0 else 0
    if relacion > 0.02:
        return 'Alta eficiencia'
    elif relacion > 0.01:
        return 'Media eficiencia'
    else:
        return 'Baja eficiencia'

# Aplicar evaluación
df_luz['eficiencia'] = df_luz.apply(evaluar_productividad, axis=1)
print("\nEficiencia en el uso de la luz:")
print(df_luz['eficiencia'].value_counts())

# Mostrar sitios más eficientes
print("\nSitios con mayor eficiencia luz-biomasa:")
sitios_eficientes = df_luz.nlargest(3, 'Biomasa')[['Sitio', 'Luz_Cantidad', 'Biomasa', 'eficiencia']]
print(sitios_eficientes)
```

## Interpretación de resultados
Escribe tus observaciones y conclusiones sobre los patrones encontrados.

```python
import pandas as pd

# Cargar todos los datos
df_biomasa = pd.read_csv("resultado_reto3_Karen.csv")
df_clima = pd.read_csv("resultados_con_clima.csv")
df_luz = pd.read_csv("resultado_reto2_santiago_final.csv")
df_resumen = pd.read_csv("tabla_resumen_clima.csv")

# Análisis integrado
print("=== ANÁLISIS INTEGRAL DEL ECOSISTEMA ===\n")

# 1. Análisis de los 20 sitios principales
datos_completos = df_biomasa.merge(df_clima[['sitio', 'clima']], on='sitio', how='left')

print("1. DISTRIBUCIÓN GENERAL:")
print(f"Total de sitios analizados: {len(datos_completos)}")
print(f"Sitios con biomasa alta: {len(datos_completos[datos_completos['Clasificación_biomasa'] == 'Alta'])}")
print(f"Sitios con biomasa media: {len(datos_completos[datos_completos['Clasificación_biomasa'] == 'Media'])}")

print("\n2. PATRONES POR CLIMA:")
for clima in datos_completos['clima'].unique():
    subset = datos_completos[datos_completos['clima'] == clima]
    avg_biomasa = subset['biomasa'].mean()
    print(f"Clima {clima}: {len(subset)} sitios, biomasa promedio: {avg_biomasa:.1f}")

print("\n3. RELACIÓN TEMPERATURA-HUMEDAD:")
correlacion = datos_completos['temperatura'].corr(datos_completos['humedad'])
print(f"Correlación temperatura-humedad: {correlacion:.2f}")

print("\n4. ANÁLISIS DE LOS SITIOS DE SANTIAGO:")
print(f"Sitios con luz alta: {len(df_luz[df_luz['categoria_luz'] == 'alta'])}")
print(f"Sitios con luz media: {len(df_luz[df_luz['categoria_luz'] == 'media'])}")
print(f"Sitios con luz baja: {len(df_luz[df_luz['categoria_luz'] == 'baja'])}")

# 5. Conclusiones generales
print("CONCLUSIONES GENERALES")

print("""
PRINCIPALES HALLAZGOS:

1. DISTRIBUCIÓN DE CLIMAS:
   - El clima 'húmedo templado' predomina en la región
   - Existe correlación negativa entre temperatura y humedad

2. BIOMASA:
   - La mayoría de sitios presentan biomasa clasificada como 'Alta'
   - Los humedales muestran consistentemente alta biomasa

3. RELACIÓN LUZ-BIOMASA:
   - Sitios con mayor luz no necesariamente tienen mayor biomasa
   - La eficiencia en el uso de la luz varía significativamente
""")

# Guardar análisis integrado
analisis_final = datos_completos.copy()
analisis_final.to_csv("analisis_integrado_ecosistema.csv", index=False)
print("✓ Análisis integrado guardado en 'analisis_integrado_ecosistema.csv'")
```
